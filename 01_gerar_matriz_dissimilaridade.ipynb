

## Lendos os dados normalizados previamente

```python
import os
import numpy as np
from IPython.display import display
import pandas as pd

DATA_BASE_PATH = "./data"

FAC_FILE = os.path.join(DATA_BASE_PATH, "mfeat-fac-norm.npy")
FOU_FILE = os.path.join(DATA_BASE_PATH, "mfeat-fou-norm.npy")
KAR_FILE = os.path.join(DATA_BASE_PATH, "mfeat-kar-norm.npy")

fac_norm = np.load(FAC_FILE)
fou_norm = np.load(FOU_FILE)
kar_norm = np.load(KAR_FILE)

print("FAC Normalized")
display(pd.DataFrame(fac_norm).head(2))

print("FOU Normalized")
display(pd.DataFrame(fou_norm).head(2))

print("KAR Normalized")
display(pd.DataFrame(kar_norm).head(2))
```

## Computando as matrizes de dissimilaridade

```python
from sklearn.metrics.pairwise import euclidean_distances

fac_dissimilarity = euclidean_distances(fac_norm, fac_norm)
fou_dissimilarity = euclidean_distances(fou_norm, fou_norm)
kar_dissimilarity = euclidean_distances(kar_norm, kar_norm)

print("FAC Dissimilarity  shape: ", fac_dissimilarity.shape)
pd.DataFrame(fac_dissimilarity).head(3)
```

```python
print("FOU Dissimilarity shape: ", fou_dissimilarity.shape)
pd.DataFrame(fou_dissimilarity).head(3)
```

```python

print("KAR Dissimilarity shape: ", kar_dissimilarity.shape)
pd.DataFrame(kar_dissimilarity).head(3)
```

## Exportando matrizes de dissimilaridades

# Nova seção

```python
np.save(FAC_FILE.replace("norm.npy","dissimilarity"), fac_dissimilarity)
np.save(FOU_FILE.replace("norm.npy","dissimilarity"), fou_dissimilarity)
np.save(KAR_FILE.replace("norm.npy","dissimilarity"), kar_dissimilarity)
```
