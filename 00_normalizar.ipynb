```
import os
import numpy as np
import pandas  as pd


DATA_BASE_PATH = "/content/drive/My Drive/Fuzzy Clustering/data"

FAC_FILE = os.path.join(DATA_BASE_PATH, "mfeat-fac")
FOU_FILE = os.path.join(DATA_BASE_PATH, "mfeat-fou")
KAR_FILE = os.path.join(DATA_BASE_PATH, "mfeat-kar")

fac = np.loadtxt(FAC_FILE, dtype=int)
fou = np.loadtxt(FOU_FILE, dtype=float)
kar = np.loadtxt(KAR_FILE, dtype=float)
print("FAC shape:", fac.shape, "\nFOU shape:", fou.shape, "\nKAR shape:", kar.shape)
```

## Normalizando os dados
### [Descrição do algoritmo Referêcia](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.minmax_scale.html#examples-using-sklearn-preprocessing-minmax-scale)



```
from sklearn.preprocessing import minmax_scale

fac = minmax_scale(fac, copy=False)
fou = minmax_scale(fou, copy=False)
kar = minmax_scale(kar, copy=False)

pd.DataFrame(fac).head(5)
```

```
pd.DataFrame(fou).head(5)
```

```
pd.DataFrame(kar).head(5)
```

## Exportando os dados normalizados

```
np.save(f"{FAC_FILE}-norm", fac)
np.save(f"{FOU_FILE}-norm", fou)
np.save(f"{KAR_FILE}-norm", kar)
```
