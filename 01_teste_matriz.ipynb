

## Lendos os dados normalizados previamente

```python
import os
import numpy as np
from IPython.display import display
import pandas as pd

DATA_BASE_PATH = "./data"

sFAC_FILE = os.path.join(DATA_BASE_PATH, "mfeat-fac-snorm.npy")
sFOU_FILE = os.path.join(DATA_BASE_PATH, "mfeat-fou-snorm.npy")
sKAR_FILE = os.path.join(DATA_BASE_PATH, "mfeat-kar-snorm.npy")

sfac_norm = np.load(sFAC_FILE)
sfou_norm = np.load(sFOU_FILE)
skar_norm = np.load(sKAR_FILE)

print("sFAC Normalized")
display(pd.DataFrame(sfac_norm).head(2))

print("sFOU Normalized")
display(pd.DataFrame(sfou_norm).head(2))

print("sKAR Normalized")
display(pd.DataFrame(skar_norm).head(2))
```

## Computando as matrizes de dissimilaridade

```python
from sklearn.metrics.pairwise import euclidean_distances

sfac_dissimilarity = euclidean_distances(sfac_norm, sfac_norm)
sfou_dissimilarity = euclidean_distances(sfou_norm, sfou_norm)
skar_dissimilarity = euclidean_distances(skar_norm, skar_norm)

print("sFAC Dissimilarity  shape: ", sfac_dissimilarity.shape)
pd.DataFrame(sfac_dissimilarity).head(3)
```

```python
print("sFOU Dissimilarity shape: ", sfou_dissimilarity.shape)
pd.DataFrame(sfou_dissimilarity).head(3)
```

```python

print("sKAR Dissimilarity shape: ", skar_dissimilarity.shape)
pd.DataFrame(skar_dissimilarity).head(3)
```

## Exportando matrizes de dissimilaridades

# Nova seção

```python
np.save(sFAC_FILE.replace("norm.npy","dissimilarity"), sfac_dissimilarity)
np.save(sFOU_FILE.replace("norm.npy","dissimilarity"), sfou_dissimilarity)
np.save(sKAR_FILE.replace("norm.npy","dissimilarity"), skar_dissimilarity)
```
